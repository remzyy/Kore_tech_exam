```python
import json
import pandas as pd
```

```python
f = open('Salesforce_contacts.json')
  
data = json.load(f)
```

```python
# Create a dictionary to store the unique email IDs
unique_emails = {}

# Loop through the data and add records with unique email IDs to the dictionary
for record in data:
    email = record['email']
    if email not in unique_emails:
        unique_emails[email] = record
    else:
        # If there is already a record with this email, check if the current record has a phone number
        if 'phonenumber' in record:
            # If the current record has a phone number, replace the existing record with it
            unique_emails[email] = record

# Create a list of the unique records with both email IDs and phone numbers
unique_records = [record for record in unique_emails.values() if 'phonenumber' in record]

# Print the unique records
for record in unique_records:
    print(record)

# Write the records with email addresses back to the file
with open('data1.json', 'w') as f:
  json.dump(unique_records, f, indent=4)
   
```

```python
import pandas as pd
import numpy as np
# Load the JSON data into a pandas DataFrame
df = pd.read_json('data1.json')

# Replace blank values with NaN
df['email'] = df['email'].replace('', np.nan)

# Drop all rows where the emailid column is null
df = df.dropna(subset=['email'])
df
```
